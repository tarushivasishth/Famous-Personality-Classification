```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import requests
import os
import time

# Provide the path to your ChromeDriver executable
chrome_driver_path = "./chromedriver.exe"  # Update this with your actual path

# Setup Selenium using the Service class
service = Service(executable_path=chrome_driver_path)
driver = webdriver.Chrome(service=service)

# Navigate to the website
driver.get('https://www.google.com/search?q=ratan+tata+images&rlz=1C1CHBF_enIN916IN916&sca_esv=18fa2256dcc18326&sca_upv=1&cs=0&udm=2&biw=1280&bih=551&sxsrf=ADLYWIKZkFGR54bhCVId6HNS0RrbQcoikg%3A1727384755011&ei=s8z1Zsw0zJax4w_crsDBCA&oq=ratan+&gs_lp=Egxnd3Mtd2l6LXNlcnAiBnJhdGFuICoCCAAyBBAjGCcyCBAAGIAEGLEDMggQABiABBixAzIFEAAYgAQyCBAAGIAEGLEDMgUQABiABDIFEAAYgAQyCBAAGIAEGLEDMgUQABiABDIFEAAYgARIzhlQtwdY7xBwAXgAkAEAmAGMAaAB6gWqAQMwLja4AQPIAQD4AQGYAgegApYGqAIKwgIHECMYJxjqAsICChAAGIAEGEMYigXCAhAQABiABBixAxhDGIMBGIoFwgINEAAYgAQYsQMYQxiKBcICDhAAGIAEGLEDGIMBGIoFwgILEAAYgAQYsQMYgwHCAgQQABgDmAMNkgcDMS42oAeWJw&sclient=gws-wiz-serp')

# Specify the directory to save images
save_directory = "./dataset/ratan_tata"  # Change this to your desired directory
os.makedirs(save_directory, exist_ok=True)  # Create directory if it doesn't exist

# Download images, limited to 100
downloaded_images = 0  # Keep track of downloaded images
img_urls_set = set()  # Use a set to avoid duplicates

while downloaded_images < 100:
    # Extract images
    for img in driver.find_elements(By.TAG_NAME, 'img'):
        # Get image dimensions
        width = img.get_attribute('width')
        height = img.get_attribute('height')
        
        # Convert width and height to integers if they are not None
        if width and height:
            width = int(width)
            height = int(height)
            
            # Filter out small images (e.g., logos might be smaller than 100x100)
            if width > 100 and height > 100:
                img_url = img.get_attribute('src')
                
                # Ensure the image URL is valid and not already downloaded
                if img_url and img_url.startswith('http') and img_url not in img_urls_set:
                    img_urls_set.add(img_url)
                    downloaded_images += 1
                    print(f"Downloading image {downloaded_images} from {img_url}...")
                    
                    try:
                        img_data = requests.get(img_url).content
                        with open(os.path.join(save_directory, f'image_{downloaded_images}.jpg'), 'wb') as img_file:
                            img_file.write(img_data)
                        print(f"Image {downloaded_images} downloaded and saved as image_{downloaded_images}.jpg")
                    except Exception as e:
                        print(f"Failed to download image {downloaded_images}: {e}")
                        
                if downloaded_images >= 100:
                    break

    # Scroll down to load more images
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Wait for new images to load

# Quit the browser
driver.quit()
```
